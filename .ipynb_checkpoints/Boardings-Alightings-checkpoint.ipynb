```python
import numpy as np
import os
import pandas as pd
```

```python
dirWork         = os.getcwd()
dirInput        = os.path.join(dirWork, r'input'       )
dirIntermediate = os.path.join(dirWork, r'intermediate')
dirResults      = os.path.join(dirWork, r'results'     )

print("Working Directory: "      + dirWork   )
print("Input Directory: "        + dirInput  )
print("Intermediate Directory: " + dirInput  )
print("Results Directory: "      + dirResults)
```

```python
#read in on-board survey

csv_OBS = os.path.join(dirInput, r'2019 Final Weighted UTA OD Data - 2021-01-27.csv')
df_OBS  = pd.read_csv(csv_OBS)

#only have non-na id rows
df_OBS = df_OBS[df_OBS['id'].notna()]

display(df_OBS)
```

```python
#list out all columns
list(df_OBS)
```

```python
boarding_stop_lat = [
                     'prev_tran_1_on_bus_lat',
                     'prev_tran_2_on_bus_lat',
                     'prev_tran_3_on_bus_lat',
                     'prev_tran_4_on_bus_lat',
                     'stop_on_lat'           ,
                     'next_tran_1_on_bus_lat',
                     'next_tran_2_on_bus_lat',
                     'next_tran_3_on_bus_lat',
                     'next_tran_4_on_bus_lat'
                    ]

boarding_stop_lon = [
                     'prev_tran_1_on_bus_long',
                     'prev_tran_2_on_bus_long',
                     'prev_tran_3_on_bus_long',
                     'prev_tran_4_on_bus_long',
                     'stop_on_long'           ,
                     'next_tran_1_on_bus_long',
                     'next_tran_2_on_bus_long',
                     'next_tran_3_on_bus_long',
                     'next_tran_4_on_bus_long'
                     ]

alighting_stop_lat = [
                     'prev_tran_1_off_bus_lat',
                     'prev_tran_2_off_bus_lat',
                     'prev_tran_3_off_bus_lat',
                     'prev_tran_4_off_bus_lat',
                     'stop_off_lat'           ,
                     'next_tran_1_off_bus_lat',
                     'next_tran_2_off_bus_lat',
                     'next_tran_3_off_bus_lat',
                     'next_tran_4_off_bus_lat'
                    ]

alighting_stop_lon = [
                     'prev_tran_1_off_bus_long',
                     'prev_tran_2_off_bus_long',
                     'prev_tran_3_off_bus_long',
                     'prev_tran_4_off_bus_long',
                     'stop_off_long'           ,
                     'next_tran_1_off_bus_long',
                     'next_tran_2_off_bus_long',
                     'next_tran_3_off_bus_long',
                     'next_tran_4_off_bus_long'
                     ]

segment_mode       = [
                      'Modefrom1',
                      'Modefrom2',
                      'Modefrom3',
                      'Modefrom4',
                      'Mode_Fin' ,
                      'Modeto1'  ,
                      'Modeto2'  ,
                      'Modeto3'  ,
                      'Modeto4'
                     ]

df_Boardings               = pd.DataFrame()
df_Boardings_LCLTripsOnly  = pd.DataFrame()
df_Alightings_LCLTripsOnly = pd.DataFrame()

#define AdjCode for pivot when model is not related to forecast
#AADT will be adjusted based on difference between scenario and forecast scenario TDMs
#if ForecastScenarioID is note defined, then the associated forecast volume will be used for that year
#DBF don't include filename extension

for index, row in df_OBS.iterrows():
    #print(row['id'])
    
    for s, segment in enumerate(segment_mode):
        if not pd.isnull(row[segment_mode[s]]):
            #print (row[segment_mode[s]])
            
            df_Boardings = df_Boardings.append({'id'         : row['id'                ],
                                                'stop_lat'   : row[boarding_stop_lat[s]],
                                                'stop_lon'   : row[boarding_stop_lon[s]],
                                                'board_mode' : row[segment_mode[s]     ],
                                                'trip_mode'  : row['Mode_Fin'          ],
                                                'weight'     : row['linked_weight_adj' ]
                                               },ignore_index=True)
            
            #trips for LCL only, no transfer to higher modes
            if row['Linked_Mode'] == 4:
                df_Boardings_LCLTripsOnly = df_Boardings_LCLTripsOnly.append({'id'         : row['id'                ],
                                                                              'stop_lat'   : row[boarding_stop_lat[s]],
                                                                              'stop_lon'   : row[boarding_stop_lon[s]],
                                                                              'weight'     : row['linked_weight_adj' ]
                                                                             },ignore_index=True)

    for s, segment in enumerate(segment_mode):
        if not pd.isnull(row[segment_mode[s]]):
            #print (row[segment_mode[s]])
                        
            #trips for LCL only, no transfer to higher modes
            if row['Linked_Mode'] == 4:
                df_Alightings_LCLTripsOnly = df_Alightings_LCLTripsOnly.append({'id'         : row['id'                 ],
                                                                                'stop_lat'   : row[alighting_stop_lat[s]],
                                                                                'stop_lon'   : row[alighting_stop_lon[s]],
                                                                                'weight'     : row['linked_weight_adj'  ]
                                                                               },ignore_index=True)


display(df_Boardings)
display(df_Boardings_LCLTripsOnly )
display(df_Alightings_LCLTripsOnly)
```

```python
display("All Boardings: ")
display(df_Boardings[['weight']].sum())
display("LCL only Boardings: ")
display(df_Boardings_LCLTripsOnly[['weight']].sum())
display("LCL only Alightings: ")
display(df_Alightings_LCLTripsOnly[['weight']].sum())
```

Note: Checks out!! Total boardings from OBS (unlinked_weight_adj) is 158,739, which matches sum of weight when listed by boarding separately.

```python
#aggregate by board mode
df_Boardings_Stop_BoardMode = df_Boardings.groupby(['stop_lat','stop_lon','board_mode'],as_index=False).agg({'weight':[np.sum]})

#rename columns
df_Boardings_Stop_BoardMode.columns = ('stop_lat','stop_lon','board_mode','boardings')

display(df_Boardings_Stop_BoardMode)
display(df_Boardings_Stop_BoardMode[['boardings']].sum())

#aggregate by trip mode
df_Boardings_Stop_TripMode  = df_Boardings.groupby(['stop_lat','stop_lon','trip_mode' ],as_index=False).agg({'weight':[np.sum]})

#rename columns
df_Boardings_Stop_TripMode.columns = ('stop_lat','stop_lon','trip_mode','boardings')

display(df_Boardings_Stop_TripMode)
display(df_Boardings_Stop_TripMode[['boardings']].sum())
```

```python
df_Boardings_Stop_BoardMode_pvt = df_Boardings_Stop_BoardMode.pivot_table(index=['stop_lat','stop_lon'], columns='board_mode')


df_Boardings_Stop_BoardMode_pvt.columns = df_Boardings_Stop_BoardMode_pvt.columns.droplevel(0)
df_Boardings_Stop_BoardMode_pvt = df_Boardings_Stop_BoardMode_pvt.reset_index()

df_Boardings_Stop_BoardMode_pvt.index.name = 'stopid'

df_Boardings_Stop_BoardMode_pvt = df_Boardings_Stop_BoardMode_pvt.rename(columns={4.0: "BoardMode_4", 5.0: "BoardMode_5", 6.0: "BoardMode_6", 7.0: "BoardMode_7", 8.0: "BoardMode_8", 9.0: "BoardMode_9"})

display(df_Boardings_Stop_BoardMode_pvt)
display(df_Boardings_Stop_BoardMode_pvt[['BoardMode_4','BoardMode_5','BoardMode_6','BoardMode_7','BoardMode_8','BoardMode_9']].sum())
display(df_Boardings_Stop_BoardMode_pvt[['BoardMode_4','BoardMode_5','BoardMode_6','BoardMode_7','BoardMode_8','BoardMode_9']].sum().sum())

#df_Boardings_Stop_BoardMode_pvt.columns
```

```python
df_Boardings_Stop_TripMode_pvt = df_Boardings_Stop_TripMode.pivot_table(index=['stop_lat','stop_lon'], columns='trip_mode')

df_Boardings_Stop_TripMode_pvt.columns = df_Boardings_Stop_TripMode_pvt.columns.droplevel(0)
df_Boardings_Stop_TripMode_pvt = df_Boardings_Stop_TripMode_pvt.reset_index()

df_Boardings_Stop_TripMode_pvt.index.name = 'stopid'

df_Boardings_Stop_TripMode_pvt = df_Boardings_Stop_TripMode_pvt.rename(columns={4.0: "TripMode_4", 5.0: "TripMode_5", 6.0: "TripMode_6", 7.0: "TripMode_7", 8.0: "TripMode_8", 9.0: "TripMode_9"})

display(df_Boardings_Stop_TripMode_pvt)
display(df_Boardings_Stop_TripMode_pvt[['TripMode_4','TripMode_5','TripMode_6','TripMode_7','TripMode_8','TripMode_9']].sum())
display(df_Boardings_Stop_TripMode_pvt[['TripMode_4','TripMode_5','TripMode_6','TripMode_7','TripMode_8','TripMode_9']].sum().sum())

#df_Boardings_Stop_TripMode_pvt.columns
```

```python
df_Boardings_All = pd.DataFrame.merge(df_Boardings_Stop_BoardMode_pvt, df_Boardings_Stop_TripMode_pvt, on=['stop_lat','stop_lon'], how='outer')
df_Boardings_All = df_Boardings_All.fillna(0)
df_Boardings_All
```

```python
df_Boardings_All.to_csv(os.path.join(dirResults, r'OBS_Stop_Boardings_All.csv'),index=False)
```

```python
#aggregate by board mode
df_Boardings_Stop_LCLTripsOnly = df_Boardings_LCLTripsOnly.groupby(['stop_lat','stop_lon'],as_index=False).agg({'weight':[np.sum]})

#rename columns
df_Boardings_Stop_LCLTripsOnly.columns = ('stop_lat','stop_lon','boardings')
display(df_Boardings_Stop_LCLTripsOnly)
display(df_Boardings_Stop_LCLTripsOnly['boardings'].sum())

#aggregate by board mode
df_Alightings_Stop_LCLTripsOnly = df_Alightings_LCLTripsOnly.groupby(['stop_lat','stop_lon'],as_index=False).agg({'weight':[np.sum]})

#rename columns
df_Alightings_Stop_LCLTripsOnly.columns = ('stop_lat','stop_lon','alightings')
display(df_Alightings_Stop_LCLTripsOnly)
display(df_Alightings_Stop_LCLTripsOnly['alightings'].sum())
```

```python
df_Boardings_Stop_LCLTripsOnly.to_csv(os.path.join(dirResults, r'OBS_Stop_Boardings_LCLTripsOnly.csv'),index=False)
df_Alightings_Stop_LCLTripsOnly.to_csv(os.path.join(dirResults, r'OBS_Stop_Boardings_LCLTripsOnly.csv'),index=False)
```

```python
df_BoardingsAndAlightings_Stop_LCLTripsOnly = pd.DataFrame.merge(df_Boardings_Stop_LCLTripsOnly, df_Alightings_Stop_LCLTripsOnly, on=('stop_lat','stop_lon'), how='outer')
df_BoardingsAndAlightings_Stop_LCLTripsOnly = df_BoardingsAndAlightings_Stop_LCLTripsOnly.fillna(0)
df_BoardingsAndAlightings_Stop_LCLTripsOnly['TotalOnOff'] = df_BoardingsAndAlightings_Stop_LCLTripsOnly['boardings'] + df_BoardingsAndAlightings_Stop_LCLTripsOnly['alightings']
df_BoardingsAndAlightings_Stop_LCLTripsOnly['TotalOnOff_Half'] = df_BoardingsAndAlightings_Stop_LCLTripsOnly['TotalOnOff']/2
display(df_BoardingsAndAlightings_Stop_LCLTripsOnly)
display(df_BoardingsAndAlightings_Stop_LCLTripsOnly['TotalOnOff_Half'].sum())
df_BoardingsAndAlightings_Stop_LCLTripsOnly.to_csv(os.path.join(dirResults, r'OBS_Stop_BoardingsAndAlightings_LCLTripsOnly.csv'),index=False)
```

# TDM Boardings and Alightings for Comparison

```python
from arcpy import env
from arcpy import TableToTable_conversion
```

```python
dbf_TDMpastation = os.path.join(dirInput       , r'from TDM Step2c3\_v832_SE19_Net19_1_PA_Station.dbf')
csv_TDMpastation = os.path.join(dirIntermediate, r'TDMPAStationData.csv')
```

```python
# Create function that converts dbfs to csvs using arcpy
def dbfToCSV(dbf, csv):
    TableToTable_conversion(dbf, os.path.dirname(csv), os.path.basename(csv))

# Allow arcpy to overwrite output files
env.overwriteOutput = True

# convert dbfs to csvs
dbfToCSV(dbf_TDMpastation, csv_TDMpastation)
```

```python
df_TDMpastation = pd.read_csv(csv_TDMpastation)
df_TDMpastation
```

```python
df_TDMBoardings_LCLOnly  = df_TDMpastation[['A','DY_D4_BRDA','DY_W4_BRDA']]
df_TDMBoardings_LCLOnly.columns = ('Node','BoardingsDr','BoardingsWk')
df_TDMAlightings_LCLOnly = df_TDMpastation[['B','DY_4_XITB' ]]
df_TDMAlightings_LCLOnly.columns = ('Node','Alightings')

display(df_TDMBoardings_LCLOnly)
display(df_TDMAlightings_LCLOnly)

df_TDMBoardings_LCLOnly['Boardings'] = df_TDMBoardings_LCLOnly['BoardingsDr'] + df_TDMBoardings_LCLOnly['BoardingsWk']


df_TDMStationBoardings_LCLOnly  = df_TDMBoardings_LCLOnly.groupby(['Node'],as_index=False).agg({'Boardings' :[np.sum]})
df_TDMStationBoardings_LCLOnly.columns = df_TDMStationBoardings_LCLOnly.columns.droplevel(1)
df_TDMStationBoardings_LCLOnly = df_TDMStationBoardings_LCLOnly[(df_TDMStationBoardings_LCLOnly['Boardings']>0)]

df_TDMStationAlightings_LCLOnly = df_TDMAlightings_LCLOnly.groupby(['Node'],as_index=False).agg({'Alightings':[np.sum]})
df_TDMStationAlightings_LCLOnly.columns = df_TDMStationAlightings_LCLOnly.columns.droplevel(1)
df_TDMStationAlightings_LCLOnly = df_TDMStationAlightings_LCLOnly[(df_TDMStationAlightings_LCLOnly['Alightings']>0)]


df_TDMStationBoardingsAndAlightings_LCLOnly = pd.DataFrame.merge(df_TDMStationBoardings_LCLOnly,df_TDMStationAlightings_LCLOnly,on='Node',how='outer')
df_TDMStationBoardingsAndAlightings_LCLOnly = df_TDMStationBoardingsAndAlightings_LCLOnly.fillna(0)

df_TDMStationBoardingsAndAlightings_LCLOnly['BoardingsAndAlightings'] = df_TDMStationBoardingsAndAlightings_LCLOnly['Boardings'] + df_TDMStationBoardingsAndAlightings_LCLOnly['Alightings']

df_TDMStationBoardingsAndAlightings_LCLOnly['BoardingsAndAlightings_Half'] = df_TDMStationBoardingsAndAlightings_LCLOnly['BoardingsAndAlightings'] / 2

df_TDMStationBoardingsAndAlightings_LCLOnly
```
